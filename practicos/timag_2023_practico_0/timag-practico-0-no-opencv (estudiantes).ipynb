### Acceso a los datos de un array  
El acceso en cada dimensión sigue el esquema [desde(incluido) : hasta(excluido) : paso]  
Si no se especifica 'desde' entonces equivale a 'desde el índice 0'  
Si no se especifica 'hasta' entonces equivale a 'hasta el último (en el sentido que dictamine 'paso')'  
El 'paso' por defecto es 1

````python
print('=================================================')
print('b')
print(b)
print('-------------------------------------------------')
# acceder a los elementos de un array
print('b[0, 0] b[0, 1] b[1, 0]')
print(b[0, 0], b[0, 1], b[1, 0])   

print('-------------------------------------------------')
# acceder a una columna de b
print('Columna 1 de b')
print(b[:,1])


print('=================================================')
print('a')
print(a)
print('-------------------------------------------------')
# acceder al primer elemento de a
print(a[0])
# acceder a los dos primeros elementos de a
print(a[:2])
# acceder al último elemento de a
print(a[-1])
# acceder a los elementos de a pero saltando de a dos
print(a[::2])
# acceder a los elementos de a pero en orden invertido
print(a[-1::-1])


print('=================================================')
print('---vista y copia---------------------------------')
print('b inicial')
print(b)
print('-------------------------------------------------')

g = b[:,1]  # g es una vista de una columna de b
print(g)    # 
print(b)
g[0] = np.random.randint(100)  # cambiamos el primer elemento de g
print(g)
print(b)    # y vemos que b también cambió ->  SON LOS MISMOS DATOS !!

h = b[:,2].copy()  # h es una copia de una columna de b. SI cambiamos h, no va a cambiar b
print(h)
h[1] = np.random.randint(100)
print(h)
print(b)
````

### Ejercicio
Dada la imagen G, se realizará el negativo de la misma:
- Implementar en este notebook
- Fiji: _Ctrl+Shift+I_ o _Edit-->Invert_

````python

# a) implementar una función negativo que recibe una imagen P (de un solo canal) y devuelve otra imagen N 
# donde se cumple que N[i,j] = 255 - P[i,j]
def negativo(P):
    # completar
    # implementar mediante dos bucles "for"
    N = np.copy(P)
    rows, cols = P.shape
    for i in range(rows):
        for j in range(cols):
            N[i,j] = 255 - P[i,j]
    return N

# b) Mostrar la imagen G y su negativo
fig, axs = plt.subplots(nrows=1, ncols=2)
axs[0].imshow(G)
axs[1].imshow(negativo(G))
plt.show()
````

### Preguntas:
a) En la función "negativo" cuál es la forma correcta de crear la imagen de salida N?  N = P o N = P.copy() ?  
b) Se puede usar la función negativo para imágenes color ? Funciona ? (ver broadcasting en numpy)  
c) Se puede sustituir la función negativo por una línea con una sola operación de numpy ?  

Respuestas:

a) N = P.copy(), puesto que de lo contrario N sería un alias de P y las modificaciones que le hagamos a N también se verán reflejadas en P. Usando el método copy obtenemos una "deep" copy en vez de una "shallow" por lo que no tendremos dicho problema.

b) Sí, es posible como en el for hacemos N[i,j] = 255 - P[i,j] y la entrada i,j de I es un vector de 3 entradas, debido al broadcasting se le sumarán 255 al opuesto de cada una de esas entradas.
$I_{i,j}=[a,b,c] => 255-[a,b,c] => [255-a,255-b,255-c]$

c) Sí, podríamos hacer : N = 255 - P. Gracias al broadcasting al escalar 255 se le "agregan dimensiones" para igualar a las del array y luego es una suma entre arrays del mismo tamaño.
$255_{1} + P_{266x400} => 255_{1x400} + P_{266x400} => 255_{266x400} + P_{266x400} => (255+P)_{266x400}$
